```python
# importing the library
import docx2txt
```

```python
# loading the resume of the applicant 
resume = docx2txt.process("My Resume-1.docx")
```

```python
# printing the document details
print(resume)
```

```python
# loading the job description document
job_desc = docx2txt.process("Software engineer job description.docx")
print(job_desc)
```

```python
# Creating a List to store resume and Job description 
text = [resume , job_desc]
```

```python
# import library to count the occurance of each word 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
count_matrix = cv.fit_transform(text)
```

```python
# Importing a library to find Similarity between the text of two documents .
from sklearn.metrics.pairwise import cosine_similarity
print(cosine_similarity(count_matrix))
```

```python
# storing % in 
match = cosine_similarity(count_matrix)[0][1]
match = match*100
match =round(match,2)
print(match)
```

```python
if(match>50):
    print("application is accepted")
else: 
    print("application is rejected")
```
